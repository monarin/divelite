```python
import numpy as np
import dask 
import dask.array as da
import xarray as xr
```

```python
arr = da.stack([da.from_delayed(dask.delayed(np.ones)(5), 
                                shape=(5,),
                                dtype=float)
                for _ in range(3)])
```

```python
arr
```

```python
xr.DataArray(arr, coords={"y": list("abc"), "x":list(range(5))})
```

```python
ds = xr.Dataset(
    {"foo": (("x", "y"), np.random.rand(4, 5))},
)
```

```python
ds.to_netcdf("/sdf/home/m/monarin/tmp/saved_on_disk.h5")
```
